# Here you can test the predicting capability of the model on unseen images downloaded from the internet or found otherwise

```python
from tensorflow.keras import models
import tensorflow as tf
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = models.load_model('drive/MyDrive/SCAII_Drive_Arjit_Agarwal_PS3/model_acc_88.5.h5')
```

```python
labels = ['Damaged Infrastructure - Eearthquake',
            'Damaged Infrastructure - Infrastructure', 
            'Fire Damage - Urban Fire', 
            'Fire Damage - Wild Fire',
            'Human Damage',
            'Land Disaster - Drought', 
            'Land Disaster - Land Slide',
            'Non Damage - Human',
            'Non Damage - Buildings street',
            'Non Damage - Wildlife forest',
            'Non Damage - Sea',
            'Water Disaster'
        ]
```

```python
img = tf.io.read_file('drive/MyDrive/SCAII_Drive_Arjit_Agarwal_PS3/test.jpg')
image = tf.io.decode_image(img, channels=3)
image = tf.image.resize(image, [224,224], method='nearest')
image = np.expand_dims(image, axis=0)
pred = model.predict(image)
prediction = np.argmax(pred, axis=-1)
print(labels[prediction[0]])
```
